```python
df = pd.read_csv('http://bit.ly/imdbratings')
df
```

```python
import pandas as pd
```

```python
df.isnull().sum()
```

```python
import ast
from collections import Counter

# Convert the string of actors to a Python list (if necessary)
df['actors_list'] = df['actors_list'].apply(ast.literal_eval)

# Flatten all actor lists into one long list
all_actors = [actor for sublist in df['actors_list'] for actor in sublist]

# Count how many times each actor appears
actor_counts = Counter(all_actors)

# Find the actor(s) who appeared the most
max_count = max(actor_counts.values())  # highest appearance count
most_actors = [actor for actor, count in actor_counts.items() if count == max_count]

print("Actors who appeared the most (", max_count, "time(s) ):")
print(most_actors)
```

```python

# See the first few rows
print(df.head())

# See info about columns, data types, and missing values
print(df.info())

# Quick statistics (only for numeric columns)
print(df.describe())
```

```python
# Check how many missing values each column has
print(df.isnull().sum())

# fill missing values with something
df.loc[:, 'genre'] = df['genre'].fillna('Unknown')
```

```python
# Check for duplicate rows
print(df.duplicated().sum())

# Remove duplicates
df = df.drop_duplicates()
```

```python
df.loc[:, 'actors_list'] = df['actors_list'].astype(str)

print(df.duplicated().sum())  # Now this will work
df = df.drop_duplicates()
```

```python
df['genre'] = df['genre'].str.strip().str.lower()
df['content_rating'] = df['content_rating'].str.strip().str.lower()
df['title'] = df['title'].str.strip()
```

```python
df
```

```python
# Example: replace inconsistent genre names
df['genre'] = df['genre'].replace({
    'sci-fi': 'science fiction',
    'rom-com': 'romance',
    'thriller/suspense': 'thriller'
})
```

```python
# Check for outliers in 'duration'
print(df['duration'].describe())

# Remove unrealistic durations
df = df[df['duration'] < 500]
```

```python
df = df.rename(columns={
    'star_rating': 'rating',
    'content_rating': 'age_rating'
})
```

```python
from collections import Counter
import ast

# Convert the actors list (which is a string) back to a Python list
df['actors_list'] = df['actors_list'].apply(ast.literal_eval)

# Flatten and count all actors
actor_counts = Counter([actor for actors in df['actors_list'] for actor in actors])
top_10_actors = actor_counts.most_common(10)

# Create a DataFrame for visualization
top_actors_df = pd.DataFrame(top_10_actors, columns=['Actor', 'Count'])

plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Actor', data=top_actors_df, palette='viridis')
plt.title("Top 10 Most Frequent Actors")
plt.xlabel("Number of Appearances")
plt.ylabel("Actor")
plt.show()
```

```python


# Convert actors_list back to list if it's a string
df['actors_list'] = df['actors_list'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Create subplot grid (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# ---------------- Plot 1: Distribution of Ratings ----------------
sns.histplot(df['rating'], bins=10, kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title("Distribution of Movie Ratings")
axes[0, 0].set_xlabel("Rating")
axes[0, 0].set_ylabel("Frequency")

# ---------------- Plot 2: Count of Movies by Genre ----------------
sns.countplot(y='genre', data=df, order=df['genre'].value_counts().index, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title("Number of Movies by Genre")
axes[0, 1].set_xlabel("Count")
axes[0, 1].set_ylabel("Genre")

# ---------------- Plot 3: Duration by Genre ----------------
sns.boxplot(x='duration', y='genre', data=df, ax=axes[1, 0], palette='coolwarm')
axes[1, 0].set_title("Movie Duration by Genre")
axes[1, 0].set_xlabel("Duration (minutes)")
axes[1, 0].set_ylabel("Genre")

# ---------------- Plot 4: Top 10 Actors ----------------
actor_counts = Counter([actor for actors in df['actors_list'] for actor in actors])
top_10_actors = actor_counts.most_common(10)
top_actors_df = pd.DataFrame(top_10_actors, columns=['Actor', 'Count'])
sns.barplot(x='Count', y='Actor', data=top_actors_df, ax=axes[1, 1], palette='magma')
axes[1, 1].set_title("Top 10 Most Frequent Actors")
axes[1, 1].set_xlabel("Appearances")
axes[1, 1].set_ylabel("Actor")

# Adjust spacing
plt.tight_layout()
plt.show()
```
